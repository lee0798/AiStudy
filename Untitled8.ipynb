```python
#data load
import seaborn as sns
```

```python
data = sns.load_dataset('penguins') # (344 row * 7 col)
```

```python
data
```

```python
data = data.dropna() #결측값 제거하기
```

```python
data
```

```python
sns.histplot(data = data, x = 'body_mass_g') # 그래프화 
```

```python
sns.histplot(data = data, x = 'body_mass_g', bins = 15) # x축 구간 갯수 정해주기 
```

```python
sns.histplot(data = data, x = 'body_mass_g', bins = 15, hue = 'species', multiple='stack')
# hue 옵션은 색에 따라 다르게 표현하라는 옵션이고 multiple을 통해 보고 나와있는 범례로만 확인가능하도록 한다.
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.title('Distribution over body_mass', fontsize = 14)
sns.histplot(data = data, x = 'body_mass_g', bins = 15, hue = 'species', multiple='stack')
plt.show()
```

```python
# 2 Distplot
sns.displot(data= data, kind = 'kde', x = 'body_mass_g', hue = 'species')
```

```python
sns.displot(data= data, kind = 'kde', x = 'body_mass_g',hue = 'species', col = 'island')
```

```python
# barplot
sns.barplot(data = data, x = 'species', y = 'body_mass_g')
```

```python
sns.barplot(data = data, x = 'body_mass_g', y = 'species', hue = 'sex')
```

```python
# countplot
sns.countplot(data = data, x = 'sex', hue = 'species')
```

```python
#boxplot
sns.boxplot(data = data, x = 'species', y = 'bill_depth_mm' )
# 1.5x IQR 규칙: 박스 플롯에서 사분위 범위(IQR)를 계산하고, IQR의 1.5배를 벗어나는 데이터 포인트를 이상치로 간주합니다. IQR은 데이터의 중앙 50% 범위를 나타내며, 1.5배를 곱한 값은 일반적으로 이상치의 경계를 정의하는 데 사용됩니다.

# 3x IQR 규칙: 보다 엄격한 기준으로 IQR의 3배를 벗어나는 데이터 포인트를 이상치로 간주하기도 합니다. 이 방법은 이상치를 더 엄격하게 식별하는 데 사용됩니다.
```

```python
sns.boxplot(data = data, x = 'species', y = 'bill_depth_mm', hue= 'sex')
#Adelie는 남여 둘다 이상치 존재
```

```python
# violineplot
sns.violinplot(data = data, x = 'species', y = 'bill_depth_mm', hue= 'sex')
```
